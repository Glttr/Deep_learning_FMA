# Plot loss/accuracy

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
logs = pd.read_csv("../outputs/logs/baseline_logs.csv")

plt.figure(figsize=(12,5))
plt.plot(logs["epoch"], logs["train_loss"], label="Train loss")
plt.plot(logs["epoch"], logs["val_loss"], label="Val loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Baseline CNN  Loss curve")
plt.legend()
plt.show()

plt.figure(figsize=(12,5))
plt.plot(logs["epoch"], logs["train_acc"], label="Train acc")
plt.plot(logs["epoch"], logs["val_acc"], label="Val acc")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Baseline CNN  Accuracy curve")
plt.legend()
plt.show()
```

```python
logs = pd.read_csv("../outputs/logs/improved_logs.csv")

plt.figure(figsize=(12,5))
plt.plot(logs["epoch"], logs["train_loss"], label="Train loss")
plt.plot(logs["epoch"], logs["val_loss"], label="Val loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("I    mproved CRNN  Loss curve")
plt.legend()
plt.show()

plt.figure(figsize=(12,5))
plt.plot(logs["epoch"], logs["train_acc"], label="Train acc")
plt.plot(logs["epoch"], logs["val_acc"], label="Val acc")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Improved CRNN  Accuracy curve")
plt.legend()
plt.show()
```
